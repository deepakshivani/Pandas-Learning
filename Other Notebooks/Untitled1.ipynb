```python
# Here first Im grouping the df by country col. So groups will be created based on unique values of country col
# Now, each group will have its own df.think of it, like a filter (df[df["country"]=="India"]), but it is not exactly a filter
# now when we apply median method to DataFrameGroupBy object, it will apply median to all group's numerical cols
# now we are selecting all rows and convertedComp col of DataFrameGroupBy obj, which will be a series

df.groupby("Country").median(numeric_only=True).loc[:,"ConvertedComp"]
```
