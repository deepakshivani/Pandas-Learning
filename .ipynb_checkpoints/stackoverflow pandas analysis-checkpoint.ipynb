```python
# Here first Im grouping the df by country col. So groups will be created based on unique values of country col
# Now, each group will have its own df.think of it, like a filter (df[df["country"]=="India"]), but it is not exactly a filter
"""
coun1 - df[df["country"]=="coun1"]
coun2 - df[df["country"]=="coun2"]
.
.
.
.
counN - df[df["country"]=="counN"]

"""
# now when we apply median method to DataFrameGroupBy object, it will apply median to all group's numerical cols
# now we are selecting all rows and convertedComp col of DataFrameGroupBy obj, which will be a series

df.groupby("Country").median(numeric_only=True).loc[:,"Convertedcomp"]

# We can also do like this, it will give the same result
# Here first Im grouping the df by country col. So groups will be created based on unique values of country col
# now we have selected only convertedcomp col which is return us a seriesgroupby object.
# seriesgroupby object is like 
"""
coun1 - (series) which will have values of all convertedcomp col, particular to country1
coun2 - (series) which will have values of all convertedcomp col, particular to country2
.
.
.
.
counN - (series) which will have values of all convertedcomp col, particular to countryN
"""
# applying median to particular series, and it will return median of particular country's convertedcomp series


df.groupby("Country")["Convertedcomp"].median(numeric_only=True)
```

# Cleaning Data - Casting Datatypes and Handling Missing Values

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
people_df=pd.DataFrame(people)
```

```python
# here, we can see that we have np.nan, None, "NA", "Missing" value. out of which, np.nan and None can be handle with fillna or dropna
# but we have to handle the "NA" and "Missing" differently

people_df
```

```python
# .isnull() method will give boolean values indicating whether the value that is passed into the argument 
# is in fact missing data, True for missing values
# notnull() is reverse of .isnull()

people_df.isnull()
```

```python
# used to drop nan values
# axis argument defines what to drop rows or cols
# how defines how to specific the dropping (any means if any of the value is nan then drop, 
# all means if all of values are nan then only drop)

# here it will drop rows if any of the cols has nan value
people_df.dropna(axis="index", how="any")
```

```python
# here it will drop rows only if all of the cols has nan value
people_df.dropna(axis="index", how="all")
```

```python
# here it will drop cols if any of the row has nan value
# below statement will drop all col bcoz, 4 row of people_df has nan values in all cols

people_df.dropna(axis="columns", how="any")
```

```python
# only drop rows when nan value is present in email col

people_df.dropna(axis="rows", how="any", subset=["email"])
```

```python
# only drop rows when nan value is present in any of email col or last col
# if we change it to all, then it will drop when nan value is present in both, email col and last col

people_df.dropna(axis="rows", how="any", subset=["email", "last"])
```

```python
# we can deal with custom null/NA values like "NA" and "Missing"
# these are string which filled by peoples as NA values

people_df.replace("NA", np.nan, inplace=True)
people_df.replace("Missing", np.nan, inplace=True)
people_df

# another way of handling custom NA values while importing csv
# create a list of custom NA values and pass it to argument na_values while reading the csv

# na_val=["NA", "Missing"]
# df= pd.read_csv("data/2019/survey_results_public.csv", na_values=na_val)
```

```python
people_df.dropna(how="all")
```

```python
# will return a df, telling if individual values are na or not

people_df.isna()
```

```python
# filling the nan value of df with some value 

people_df.fillna("MISSING")
```

```python
# fillin nan value of series with some value. we can use some statitics here for numerical data to decide a value and fill it

people_df["age"].fillna(25)
```

```python
# this will throw error as all values are not int or float for mathematical calculation.
# see the dict at the first cell after heading, this col has string values

# people_df["age"].mean()
```

```python
# when you try to converting a col dtypes that has nan values, then use float dtype as nan is itself a float dtype value
# Nan values are going to stay nan, but we can now do the calculation on that col

people_df["age"]=people_df["age"].astype(float)
people_df["age"].mean()
```

```python
# get average years of coding from stackoverflow dataset
# here if we fo, df["Yearscode"].unique(), then we got to them some values are string line like "Less than 1 year" and "More than 50 years", 
# which we will replace with proper values then cast all values to float and get the mean

df["Yearscode"].replace("Less than 1 year", 0, inplace=True)
df["Yearscode"].replace("More than 50 years", 50, inplace=True)
df["Yearscode"]=df["Yearscode"].astype(float)
df["Yearscode"].mean()
```

```python
# To get the aggregated values for different time frame say day or week, then we use resampling
# Resampling is kind of similar like grouping where we split/break the data based on some time basis(day, week)
# and then apply a function to get back a aggregated value for that period of time

# get the code to specific the resample time limit here, https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects

ethDf["High"].resample("D").max()

# when you resample the data say on Day basis, it will group the data per day basis, like
# """
# day1 -series(121,112,142,......)
# .
# .
# dayN -series (hourly values)

# """
# then apply the max function to get the max value of particular day's grouped series
```

```python
# use matplotlib to plot the high of dataset
%matplotlib inline
ethDf["High"].resample("D").max().plot()
```

```python
# Resampling muliiple cols at once
# resample the df based on your requirment
# then apply agg method to define which aggregation method (max, min, get_close) to apply on what col, with the help of dict
# here i created a custom agg function get_close as well, which is applied to close col

def get_close(x):
    return x.iloc[-1]


ethDf.resample("W").agg({"Close":get_close, "High":"max", "Low":"min"})
```

```python
# this function will change the format of list of dict to a main dict
# from this = [{"id":1, "name":"sam"}, {"id":2, "name":"ram"}, {"id":3, "name":"fam"}]
# to this = {"id":[1,2,3], "name":["sam", "ram", "fam"]}

def one_dict(list_dict):
    keys=list_dict[0].keys()
    out_dict={key:[] for key in keys}
    for dict_ in list_dict:
        for key, value in dict_.items():
            out_dict[key].append(value)
    return out_dict
```
